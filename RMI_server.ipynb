```python
# SERVER CELL

import Pyro4
import threading

@Pyro4.expose
class StringConcatenator:
    def concatenate(self, str1, str2):
        print(f"Received: {str1}, {str2}")
        return str1 + str2

def start_server():
    daemon = Pyro4.Daemon(host="localhost")  # Bind to localhost
    uri = daemon.register(StringConcatenator)
    print("Server running. URI:", uri)
    
    # Run the server loop in a separate thread
    daemon.requestLoop()

# Run the server in a background thread
thread = threading.Thread(target=start_server)
thread.daemon = True
thread.start()
```

1. What is Remote Method Invocation (RMI)?
Answer:
RMI lets a program call methods on objects that are on a different computer. It makes it possible to use methods from a remote server just like they are local.

2. What is Pyro4 in Python?
Answer:
Pyro4 is a Python library that allows programs to communicate with each other over a network. It makes it easy for one program to use methods from another program running on a different computer.

3. How does Pyro4 make a method available for remote use?
Answer:
Pyro4 uses a special marker, @Pyro4.expose, to make a method accessible from other computers. This tells Pyro4, "Hey, this method can be called remotely."

4. What does Pyro4.Daemon do on the server side?
Answer:
Pyro4.Daemon creates a server that listens for requests from clients. It allows the server to accept and respond to calls for remote methods.

5. What is a Proxy in Pyro4?
Answer:
A Proxy is like a "remote control" for a server object. The client uses it to call methods on the server as if it’s a local object, but Pyro4 handles sending the request over the network.

6. What does the URI in Pyro4 mean?
Answer:
A URI is a unique address that tells the client where to find the server and which object to interact with. It’s like a web address for remote objects.

7. How does Pyro4 communicate between the client and the server?
Answer:
Pyro4 sends method calls and data from the client to the server over the network. It turns the data into a format that can travel through the network, then turns it back into usable data on the server.

8. Why do we use threads in a Pyro4 server?
Answer:
Threads allow the server to handle multiple clients at the same time without waiting for one to finish before starting the next. This makes the server faster and more responsive.
